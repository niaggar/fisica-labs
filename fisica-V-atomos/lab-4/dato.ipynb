```python
max_angle = np.array([])
max_intensity = np.array([])
init_angle = np.array([])


for i in range(0, len(voltaje)):
    intensidad_lista = df['i_' + str(i + 1)].dropna()
    angulo_lista = df['a_' + str(i + 1)].dropna() + ang_init

    # Get the index of the maximum element from a Numpy array
    index = np.where(intensidad_lista == max(intensidad_lista))

    max_angle = np.append(max_angle, angulo_lista[index[0][0]])
    max_intensity = np.append(max_intensity, max(intensidad_lista))

    last_angle = angulo_lista[len(angulo_lista) - 1]
    init_angle = np.append(init_angle, last_angle)

print('Intensidad maxima: ', max_intensity)
print('Angulos donde se da intensidad maxima: ', max_angle)
print('Angulos iniciales: ', init_angle * 18)
plt.plot(voltaje, max_angle, 'o', color='red', label = 'V = 4.0 $V$')
plt.xlabel('Voltaje $[V]$', fontdict = { 'color': 'black', 'size': 20 })
plt.ylabel('Angulo $[^{\circ}]$', fontdict = { 'color': 'black', 'size': 20 })
plt.show()
```

```python
def lamd(r):
    ratio = 1/59.740
    A = 13915.59
    B = 1.689

    n = np.sqrt(((2 / np.sqrt(3)) * np.sin((62.8 - r) * (np.pi / 180)) + (1/2))**2 + (3/4))
    lamd = np.sqrt( (A) / (n - B))
    
    return lamd



def Temp(I,V):
    R = V/I
    T1 = 145.15105 + 161.75397*R - 0.99351*R**2
    T2 = 300 + ( R/0.84 - 1)/0.0045
    return T1

def Real(T):
    return 1960513.4962938044 / T


print(max_angle)

t = Temp(corriente, voltaje)
l = lamd(max_angle)

print('Temperatura: ', t)
print('Longitud de onda: ', l)

plt.plot(t, l, 'o', color='red')
plt.plot(t, Real(t), '-', color='blue')
plt.xlabel('Temperatura $[K]$', fontdict = { 'color': 'black', 'size': 20 })
plt.ylabel('Longitud de onda $[nm]$', fontdict = { 'color': 'black', 'size': 20 })
plt.show()
```
